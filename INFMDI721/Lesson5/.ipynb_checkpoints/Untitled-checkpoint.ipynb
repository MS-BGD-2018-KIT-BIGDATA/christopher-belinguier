## Densité de médecin pour 100.000 habitants par spécialité

```python
# Load data
rpps_medecins = pd.read_csv('rpps_medecins.csv', sep=';', na_values=['?', ''], encoding='latin1', header=4) #skiprows=4,

# Supprime toutes les lignes ayant des valeurs manquantes
len_rpps_medecins_init = len(rpps_medecins)
rpps_medecins = rpps_medecins.dropna(axis=0, how='any')
len_rpps_medecins_clean = len(rpps_medecins)

# Affiche le nouveau nombre de ligne supprimé apres "nettoyage"
print("Nombre de lignes supprimés : " + str(len_rpps_medecins_init - len_rpps_medecins_clean))
rpps_medecins.head()
```
